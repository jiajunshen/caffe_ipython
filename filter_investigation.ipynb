```python
import caffe
import os
import deepdish as dd
import amitgroup.plot as gr
os.chdir("/Users/jiajunshen/Documents/Research/caffe/")
caffe.set_mode_gpu()
solver = caffe.get_solver("./examples/mnist/lenet_solver.prototxt")
```

```python
from amitgroup import io
allData = []
for i in range(10):
    allData.append(io.load_mnist(dataset='training',digits = [i],return_labels=False))
```

```python
solver.net.blobs['conv2'].diff.shape
```

```python
allData[0].shape
```

```python
listConvGradient = [[] for i in range(10)]
listIP2Gradient = [[] for i in range(10)]
listConvActivation = [[] for i in range(10)]
listIP2Activation = [[] for i in range(10)]


for i in range(1000):
    solver.step(1)
    #print i
    if i>=10 and i <990:
        continue
    for j in range(10):
        convArray = None
        ip2Array = None
        convGradientArray = None
        ip2GradientArray = None
        for k in range(10):
            solver.net.blobs['data'].data[...] = allData[j][64 * k:64*(k + 1)].reshape(64, 1, 28, 28)
            solver.net.blobs['label'].data[...] = np.ones(64) * j
            solver.net.forward(start = 'conv2')
            solver.net.backward()
            #The activation
            if convArray == None:
                convArray = solver.net.blobs['conv2'].data
                ip2Array = solver.net.blobs['ip2'].data
                convGradientArray = solver.net.blobs['conv2'].diff
                ip2GradientArray = solver.net.blobs['ip1'].diff
            else:
                convArray = np.concatenate((convArray, solver.net.blobs['conv2'].data))
                ip2Array = np.concatenate((ip2Array, solver.net.blobs['ip2'].data))
                convGradientArray = np.concatenate((convGradientArray, solver.net.blobs['conv2'].diff))
                ip2GradientArray = np.concatenate((ip2GradientArray, solver.net.blobs['ip1'].diff))
        listConvActivation[j].append(convArray)
        listIP2Activation[j].append(ip2Array)
        listConvGradient[j].append(convGradientArray)
        listIP2Gradient[j].append(ip2GradientArray)
```

```python
np.save("/Users/jiajunshen/Desktop/filter_investigation_activation.npy",listConvActivation)
```
