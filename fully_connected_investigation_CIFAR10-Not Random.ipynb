```python
#randomly divide the entire label sets into to classes.
def newPartition(trainingLabel, testLabel):
    sampleCluster1 = np.random.randint(2, 4)
    sampleCluster2 = np.random.randint(2, 4)
    group1 = np.random.choice(range(10),size = sampleCluster1,replace=False)
    group2 = np.random.choice([x for x in range(10) if x not in group1],size = sampleCluster2,replace=False)
    newTestLabel = [change(x, group1, group2) for x in testLabel]
    newTrainLabel = [change(x, group1, group2) for x in trainingLabel]
    testingDataIndex = np.array([x in group1 or x in group2 for x in testLabel])
    trainingDataIndex = np.array([x in group1 or x in group2 for x in trainingLabel])
    return newTrainLabel, newTestLabel, group1, group2, testingDataIndex, trainingDataIndex
```

```python
activationArray = None
for i in range(500):
    print i
    solver.net.blobs['data'].data[...] = np.array(trainingData[i * 100: (i + 1) * 100].reshape(100,3,32,32))
    solver.net.forward(start = "conv1")
    if activationArray == None:
        activationArray = np.array(solver.net.blobs['pool3'].data)
        tem = np.array(solver.net.blobs['pool3'].data)
    else:
        activationArray = np.concatenate((activationArray, np.array(solver.net.blobs['pool3'].data)))
        
    
```

```python
activationTestArray = None

for i in range(100):
    print i
    solver.net.blobs['data'].data[...] =np.array(testingData[i * 100: (i + 1) * 100].reshape(100,3,32,32))
    solver.net.forward(start = "conv1")
    if activationTestArray == None:
        activationTestArray = np.array(solver.net.blobs['pool3'].data)
        tmp = np.array(solver.net.blobs['pool3'].data)
    else:
        activationTestArray = np.concatenate((activationTestArray, np.array(solver.net.blobs['pool3'].data)))
```

```python
#generate 500 labels for the training and the testing
alteredTrainingLabels = []
alteredTestingLabels = []
testingDataIndexList = []
trainingDataIndexList = []
sampledPartition = []
for i in range(100):
    newTrainLabel, newTestLabel, group1, group2, testingDataIndex, trainingDataIndex = newPartition(trainingDataLabel, testingDataLabel)
    alteredTrainingLabels.append(newTrainLabel)
    alteredTestingLabels.append(newTestLabel)
    sampledPartition.append(tuple((group1, group2)))
    testingDataIndexList.append(testingDataIndex)
    trainingDataIndexList.append(trainingDataIndex)
```

```python
np.mean(np.array(alteredTrainingLabels[0])[trainingDataIndexList[0]])
```

```python
##generate SVM
from sklearn import svm
allSVM = []
predictionResult = []
for i in range(100):
    print i
    clf = svm.LinearSVC()
    trainingDataSize = activationArray[trainingDataIndexList[i]].shape[0]
    testingDataSize = activationTestArray[testingDataIndexList[i]].shape[0]
    clf.fit(activationArray[trainingDataIndexList[i]].reshape(trainingDataSize,-1),np.array(alteredTrainingLabels[i])[trainingDataIndexList[i]])
    allSVM.append(clf)
    predictionResult.append(np.mean(clf.predict(activationTestArray[testingDataIndexList[i]].reshape(testingDataSize,-1)) == np.array(alteredTestingLabels[i])[testingDataIndexList[i]]))
```

```python
predictionResult
```

```python
allCoef = []
for i in range(100):
    allCoef.append(allSVM[i].coef_)
allCoef = np.array(allCoef).reshape((100,-1))
```

```python
allBias = []
for i in range(100):
    allBias.append(allSVM[i].intercept_)
allBias = np.array(allBias).reshape((100,-1))
```
