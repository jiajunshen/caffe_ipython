```python
for i in range(0,1000):
    print i
    if i>=10 and i <990:
        continue
    for j in range(10):
        conv1GradientArray = None
        conv2GradientArray = None
        conv3GradientArray = None
        ip2GradientArray = None
        for k in range(10):
            solver.net.blobs['data'].data[...] = allData[j][100 * k:100*(k + 1)].reshape(100, 3, 32, 32)
            solver.net.blobs['label'].data[...] = np.ones(100) * j
            solver.net.forward(start = 'conv1')
            solver.net.backward()
            #The activation
            if conv1GradientArray == None:
                #conv1Array = np.array(solver.net.blobs['conv1'].data)
                #conv2Array = np.array(solver.net.blobs['conv2'].data)
                #conv3Array = np.array(solver.net.blobs['conv3'].data)
                #ip2Array = np.array(solver.net.blobs['ip1'].data)
                conv1GradientArray = np.array(solver.net.blobs['conv1'].diff)
                conv2GradientArray = np.array(solver.net.blobs['conv2'].diff)
                conv3GradientArray = np.array(solver.net.blobs['conv3'].diff)
                ip2GradientArray = np.array(solver.net.blobs['ip1'].diff)
            else:
                #conv1Array = np.concatenate((conv1Array, np.array(solver.net.blobs['conv1'].data)))
                #conv2Array = np.concatenate((conv2Array, np.array(solver.net.blobs['conv2'].data)))
                #conv3Array = np.concatenate((conv3Array, np.array(solver.net.blobs['conv3'].data)))
                #ip2Array = np.concatenate((ip2Array, np.array(solver.net.blobs['ip1'].data)))
                conv1GradientArray = np.concatenate((conv1GradientArray, np.array(solver.net.blobs['conv1'].diff)))
                conv2GradientArray = np.concatenate((conv2GradientArray, np.array(solver.net.blobs['conv2'].diff)))
                conv3GradientArray = np.concatenate((conv3GradientArray, np.array(solver.net.blobs['conv3'].diff)))
                ip2GradientArray = np.concatenate((ip2GradientArray, np.array(solver.net.blobs['ip1'].diff)))
        #listConv1Activation[j].append(conv1Array)
        #listConv2Activation[j].append(conv2Array)
        #listConv3Activation[j].append(conv3Array)
        #listIP2Activation[j].append(ip2Array)
        listConv1Gradient[j].append(conv1GradientArray)
        listConv2Gradient[j].append(conv2GradientArray)
        listConv3Gradient[j].append(conv3GradientArray)
        listIP2Gradient[j].append(ip2GradientArray)
    solver.step(1)
```
