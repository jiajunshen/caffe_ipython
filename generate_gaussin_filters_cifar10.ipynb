```python
#prepare the CIFAR-10 DATA, with the mean substracted
from amitgroup import io
data,label = io.load_cifar10("training")
data = data * 255
dataMean = np.mean(data, axis = 0)
data = data - dataMean
```

```python
def generatePatches(data):
    finalPatches = []
    for i in range(data.shape[0]):
        counter = 0
        sampleData = zero_pad(data[i],2)
        while(counter<40):
            x = np.random.randint(low = 0, high=32)
            y = np.random.randint(low = 0, high=32)
            finalPatches.append(sampleData[x:x+5,y:y+5,:])
            counter+=1
    return finalPatches
```

```python
data.shape
```

```python
def zero_pad(data, pad_size):
    data_shape = data.shape
    result_data = np.zeros((data_shape[0] + 2 * pad_size, data_shape[1] + 2 * pad_size, data_shape[2]))
    result_data[pad_size:pad_size + data_shape[0], pad_size:pad_size+data_shape[1],:] = data
    return result_data
```
